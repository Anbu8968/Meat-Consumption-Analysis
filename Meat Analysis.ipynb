```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as ans
```

```python
data=pd.read_csv("meat_consumption.csv")
```

```python
data_year=data.groupby(["time"]).sum()["value"].round(2).reset_index()
```

```python
data_year.sort_values(by="value",ascending=False,inplace=True)
```

```python
data_year
```

```python
data_year["time"]=data_year["time"].astype(str)
```

```python
data_year.head()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x="time",y="value",data=data_year)
```

```python
data_location=data.groupby(["location"]).sum()["value"].round(2).reset_index()
```
